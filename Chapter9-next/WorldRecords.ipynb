```python
URL = 'https://en.wikipedia.org/wiki/List_of_world_records_in_swimming'
```

```python
import gazpacho

html = gazpacho.get(URL)
soup = gazpacho.Soup(html)

type(soup)
```

```python
tables = soup.find('table', mode='list')

for i, table in enumerate(tables):
    rows = table.find('tr', mode='list')
    cols = rows[-1].find('td', mode='list')
    print(f'Table {i+1} has {len(rows)} rows and {len(cols)} columns')

```

```python
table = tables[0]
# rows = table.find('tr', mode='list')[1:]
# len(rows)
for i, row in enumerate(table.find('tr', mode='list')[1:]):
    columns = row.find('td', mode='list')
    event = columns[0].text
    time = columns[1].text
    # print(f'{event} - {time}')
```

```python
RECORDS = (0, 1, 3, 4)
COURSES = ('LC Men', 'LC Women', 'SC Men', 'SC Women')

# list(zip(RECORDS, COURSES))
result = {}
for table, course in zip(RECORDS, COURSES):
    # print(course)
    result[course] = {}
    for i, row in enumerate(tables[table].find('tr', mode='list')[1:]):
        columns = row.find('td', mode='list')
        event = columns[0].text
        time = columns[1].text
        if 'relay' not in event:
            result[course][event] = time

    # print()

print(result)
print(result.keys())
```

```python
print(result['LC Men'])
```

```python
conversions={
    'Free': 'freestyle',
    'Back': 'backstroke',
    'Breast': 'breaststroke',
    'Fly': 'butterfly',
    'IM': 'individual medley'
}
```

```python
event = 'Abi-10-50m-Back.txt'
*_, distance, stroke = event.removesuffix('.txt').split('-')
lookup = f'{distance} {conversions[stroke]}'
lookup
```

```python
for course in COURSES:
    print(result[course][lookup])
```

```python
import json

with open('records.json', 'w') as file:
    json.dump(result, file)
```
